# **Numpy - Pandas - Matplotlib -  Seaborn**

Para trabajar con datos en Python, estas cuatro librerías forman el "ecosistema estándar". Aunque a veces sus funciones se solapan, cada una tiene un rol específico en el flujo de trabajo de un Analista o Científico de Datos.

## **1. NumPy (El motor numérico)**

Es la base de todo. Si Python fuera un auto, NumPy sería el motor. Proporciona estructuras de datos de alto rendimiento (arrays) que son mucho más rápidas que las listas comunes de Python.

- **Para qué sirve:** Operaciones matemáticas complejas, álgebra lineal y manejo de grandes volúmenes de números.

- **En la vida real:** Se usa para cálculos científicos, procesamiento de señales y es la base sobre la que se construyen todas las demás librerías de esta lista.

## **2. Pandas (La navaja suiza de los datos)**

Es la librería para manipular tablas (llamadas DataFrames). Si estás acostumbrado a Excel, Pandas es como un Excel superpotenciado mediante código.

- **Para qué sirve:** Cargar archivos (CSV, Excel, SQL), limpiar datos faltantes, filtrar filas, agrupar información y realizar estadísticas rápidas.

- **En la vida real:** Es la herramienta principal para la fase de "Limpieza de Datos" (Data Wrangling) antes de hacer cualquier gráfico.

## **3. Matplotlib (El lienzo del artista)**

Es la librería de visualización más antigua y robusta de Python. Te da un control total (y a veces complejo) sobre cada detalle del gráfico.

- **Para qué sirve:** Crear gráficos desde cero (líneas, barras, dispersión). Puedes modificar cada eje, color, etiqueta o fuente.

- **En la vida real:** Se usa cuando necesitas un gráfico muy específico o personalizado que requiere un ajuste fino de los elementos visuales.

## **4. Seaborn (El diseñador de interiores)**

Construida sobre Matplotlib, Seaborn está diseñada para hacer gráficos estadísticos atractivos con muy poco código.

- **Para qué sirve:** Crear visualizaciones complejas de forma sencilla (como mapas de calor o distribuciones estadísticas) con una estética moderna por defecto.

- **En la vida real:** Se usa para la Exploración de Datos (EDA), ya que permite ver patrones y correlaciones de forma visualmente agradable y rápida sin pelearse con el código base de Matplotlib.

## **¿Cómo trabajan juntas? (El flujo de trabajo)**

Imagina que quieres analizar las ventas de una tienda:

**1. Pandas:** Cargas el archivo de ventas y eliminas los datos que están vacíos.

**2. NumPy:** Calculas la media móvil o aplicas una fórmula matemática a los precios.

**3. Seaborn:** Creas rápidamente un gráfico para ver la relación entre publicidad y ventas.

**4. Matplotlib:** Ajustas el tamaño final del gráfico, le pones el logo de la empresa y guardas la imagen en alta resolución.

## **¿Por qué esta combinación es tan potente?**

1. NumPy le da a Pandas la velocidad para manejar millones de filas.

2. Pandas organiza los datos para que Matplotlib sepa qué poner en el eje X y qué en el eje Y.

3. Seaborn toma los objetos de Matplotlib y les aplica capas de diseño profesional automáticamente.

## **TIPOS DE GRAFICOS**

|Gráfico	|Uso en la Vida Real|
|:-:|:-:|
|Line Chart	|Seguimiento de tendencias temporales (ej. precio de acciones, cambios de temperatura durante un mes).|
|Bar Chart	|Comparación de cantidades entre categorías discretas (ej. ventas por región, número de empleados por departamento).|
|Pie Chart	|Mostrar la composición de un total, ideal para pocas categorías (ej. cuota de mercado de 3 empresas principales).|
|Histogram	|Analizar la distribución de una variable continua y ver su frecuencia (ej. distribución de edades en una población).|
|Scatter Plot	|Identificar la relación o correlación entre dos variables numéricas (ej. relación entre años de experiencia y salario).|
|Box Plot	|Visualizar estadísticas clave (mediana, cuartiles) y detectar valores atípicos (outliers).|
|Violin Plot	|Similar al Box Plot, pero muestra además la densidad de probabilidad de los datos (útil para ver si hay múltiples picos).|
|Heatmap	|Visualizar matrices de datos, muy usado para ver la correlación entre variables o intensidad en mapas de calor.|
|Area Chart	|Similar al de líneas, pero enfatiza la magnitud del cambio a lo largo del tiempo.|
|Radar Chart	|Comparar múltiples variables cuantitativas para un solo elemento (ej. habilidades de un jugador en un videojuego).|

## **Ejemplos de principales tipos de gráficos, su utilidad en la vida real e implementación.**

```python
# 1. Procesamiento numérico avanzado
import numpy as np

# 2. Manipulación y análisis de datos (Tablas/DataFrames)
import pandas as pd

# 3. La librería base de visualización
import matplotlib.pyplot as plt

# 4. Visualización estadística de alto nivel (basada en Matplotlib)
import seaborn as sns

# Configuración opcional para que los gráficos se vean mejor en el Notebook
%matplotlib inline 
sns.set_theme(style="whitegrid") # Aplica un estilo visual limpio de Seaborn a todos los gráficos
plt.rcParams['figure.figsize'] = (10, 6) # Define un tamaño estándar para tus gráficos
```

# **1. Line Chart (Gráfico de Líneas)**

Uso en la vida real: Seguimiento de cambios en periodos de tiempo (Series Temporales). 

- **Ejemplo:** Evolución del precio del Bitcoin o temperatura diaria.

```python
# LINE CHART

x = [1, 2, 3, 4, 5]
y= [10, 12, 5, 8, 3]

plt.plot(x,y)
plt.title('Line Chart')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.show()
```

# **2. Bar Chart & Variaciones**

Uso en la vida real: Comparar cantidades entre grupos.

- **Barras Horizontales:** Útiles cuando los nombres de las categorías son largos.

- **Barras Agrupadas:** Comparar métricas de dos grupos distintos (Ej: Ventas 2023 vs 2024).

- **Barras Apiladas (Stacked):** Ver la composición de un total (Ej: Ventas totales divididas por tipo de producto).

```python
# BAR CHART

categories = ['A', 'B', 'C', 'D']
values= [25, 40, 30, 20]

plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

# **3. Pie Chart (Gráfico de Pastel o Grafico Circular)**

Uso en la vida real: Mostrar proporciones respecto a un total. 

- **Consejo:** Úsalo solo cuando tengas pocas categorías para que sea legible.

```python
# PIE CHART

labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

# **4. Histogram (Histograma)**

Uso en la vida real: Representar la distribución de una variable continua. 

- **Ejemplo:** ¿En qué rango de edad se encuentran la mayoría de mis clientes?

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frecuency')
plt.show()
```

# **5. Scatter Plot (Diagrama de Dispersion)**

El diagrama de dispersión grafica pares de datos numéricos, con una variable en cada eje, para buscar una relación entre ellos. Si las variables están correlacionadas, los puntos se ubicarán a lo largo de una línea o curva.

Uso en la vida real: se utilizan para mostrar relaciones. Para la correlación, los gráficos de dispersión ayudan a mostrar la fuerza de la relación lineal entre dos variables. 

- **Ejemplos:** 

- Marketing y Ventas: Comparar el gasto en publicidad (eje X) frente a los ingresos por ventas (eje Y) para medir la efectividad de campañas.

- Recursos Humanos: Analizar la relación entre las horas de capacitación recibidas (eje X) y el número de errores en el trabajo (eje Y).

- Producción y Calidad: Evaluar la relación entre la temperatura de una máquina (eje X) y el número de defectos en el producto final (eje Y).

```python
# SCATTER PLOT

x = np.random.rand(50)
y = 2* x + 1 + 0.1 * np.random.randn(50)

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.show()
```

# **6. Box Plot - Diagrama de Caja y Bigotes**

Uso en la vida real: Análisis estadístico profundo. 

El Box Plot muestra la distribución, mediana, cuartiles y valores atípicos (outliers) de datos numéricos

- **Ejemplos:** 

Comparación de grupos (Análisis de rendimiento): Comparar la distribución de las puntuaciones de exámenes entre diferentes escuelas, aulas o el rendimiento de empleados en diferentes equipos.

Análisis de procesos (Antes y después): Evaluar el impacto de un cambio de proceso, por ejemplo, comparando los tiempos de fabricación o de espera de atención al cliente antes y después de una optimización.

Identificación de Valores Atípicos (Outliers): Detectar anomalías en datos financieros (ej. fraudes o gastos inusualmente altos) o en mediciones de calidad de producción.

```python
# BOX PLOT

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

sns.boxplot(data=data)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

# **7. Violin Plot**

Uso en la vida real: añade la densidad de los datos (dónde se concentran más).

El Violin Plot es excelente para visualizar la distribución, densidad y forma de datos numéricos a través de diferentes categorías, combinando un diagrama de caja con la estimación de densidad kernel.

- **Ejemplos:** 

Comparación de distribuciones entre grupos: Analizar la distribución de los precios de automóviles según el fabricante o comparar las calificaciones de estudiantes entre diferentes escuelas.

Análisis de secuenciación unicelular: En biología, se utilizan para mostrar la distribución de los niveles de expresión génica en distintas poblaciones celulares.

Evaluación de modelos de Machine Learning: Comparar los valores reales contra los predichos para identificar sesgos, variaciones y el rendimiento del modelo.

Análisis de datos de tiempos y flujos: Comparar tiempos de respuesta de servicios o flujos de trabajo (kanban) para detectar cuellos de botella.

```python
# VIOLIN PLOT

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

sns.violinplot(data=data)
plt.title('Violin Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()

```

# **8. Heatmap - Graficos de Calor**

Los gráficos de calor o heatmaps son herramientas de visualización de datos que utilizan colores para representar la intensidad de valores en un área o conjunto de datos complejo.

Uso en la vida real:

**Radar:** Comparar perfiles (Ej: Atributos de un jugador: Velocidad, Fuerza, Pase).

**Area:** Similar a líneas pero enfatiza el volumen total bajo la curva.

- **Ejemplos:** 

**1. Marketing y Experiencia de Usuario (UX)**

Es el uso más común para optimizar sitios web y aplicaciones. 

Mapas de Clics: Permiten identificar qué botones o enlaces reciben más atención y cuáles son ignorados, ayudando a mejorar las llamadas a la acción (CTA).

Mapas de Desplazamiento (Scroll): Muestran hasta qué punto de la página bajan los usuarios, ideal para saber si el contenido importante queda "enterrado".

Mapas de Movimiento: Rastrean el movimiento del cursor para inferir hacia dónde se dirige la mirada del usuario. 

**2. Finanzas y Economía**
Ayudan a procesar grandes volúmenes de datos numéricos de un vistazo. 

Mercado de Valores: Los inversores usan Mapas de calor de acciones de TradingView para visualizar el rendimiento de cientos de activos simultáneamente, donde el verde suele indicar subidas y el rojo bajadas.

Análisis de Riesgos: Las empresas crean mapas de calor de riesgos para priorizar amenazas basándose en su probabilidad e impacto potencial. 

**3. Ciencia y Salud**

Biología Molecular: Se utilizan en Nature Methods para visualizar la expresión génica a través de múltiples muestras, facilitando la detección de patrones biológicos complejos.

Epidemiología: Durante crisis sanitarias, se emplean para identificar zonas geográficas con mayor concentración de casos de enfermedades. 


```python
# HEAT MAP

data = np.random.rand(10, 10)

sns.heatmap(data, annot=True)
plt.title('Heatmap')
plt.show()
```

# **9. Area Chart**

Los gráficos de área son una evolución de los gráficos de líneas. Su característica principal es que el espacio entre la línea y el eje horizontal se rellena con color, lo que ayuda a visualizar no solo la tendencia, sino también el volumen o la magnitud total de los datos.

Uso en la vida real:

**1. Seguimiento de Ventas e Ingresos**

Es el uso más clásico para mostrar el crecimiento financiero a largo plazo.

Volumen de Ventas: A diferencia de una línea simple, el área sombreada enfatiza la "masa" de dinero generada, lo que da una sensación 
visual de solidez o pérdida de mercado más clara que un punto en el espacio.

Ingresos Mensuales: Ideal para comparar el presupuesto proyectado frente al ingreso real.

**2. Composición de Categorías (Gráfico de Área Apilada)**

Este es uno de los usos más potentes, ya que permite ver cómo contribuye cada parte al total. 

Fuentes de Tráfico Web: Puedes ver el total de visitas de un sitio y, al mismo tiempo, qué parte corresponde a tráfico orgánico, directo o redes sociales.

Ventas por Región: Mostrar las ventas totales de una empresa mientras se visualiza cuánto aporta cada ciudad o país al volumen global. 


```python
# AREA CHART

x= [1, 2, 3, 4, 5]
y1 = [10, 15, 25, 30, 35]
y2 = [5, 10, 20, 25, 30]

plt.fill_between(x, y1, y2, color='skyblue', alpha= 0.4)
plt.title('Area Chart')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.show()
```
