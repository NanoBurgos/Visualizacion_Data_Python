# **Numpy - Pandas - Matplotlib -  Seaborn**

Para trabajar con datos en Python, estas cuatro librerías forman el "ecosistema estándar". Aunque a veces sus funciones se solapan, cada una tiene un rol específico en el flujo de trabajo de un Analista o Científico de Datos.

## **1. NumPy (El motor numérico)**

Es la base de todo. Si Python fuera un auto, NumPy sería el motor. Proporciona estructuras de datos de alto rendimiento (arrays) que son mucho más rápidas que las listas comunes de Python.

- **Para qué sirve:** Operaciones matemáticas complejas, álgebra lineal y manejo de grandes volúmenes de números.

- **En la vida real:** Se usa para cálculos científicos, procesamiento de señales y es la base sobre la que se construyen todas las demás librerías de esta lista.

## **2. Pandas (La navaja suiza de los datos)**

Es la librería para manipular tablas (llamadas DataFrames). Si estás acostumbrado a Excel, Pandas es como un Excel superpotenciado mediante código.

- **Para qué sirve:** Cargar archivos (CSV, Excel, SQL), limpiar datos faltantes, filtrar filas, agrupar información y realizar estadísticas rápidas.

- **En la vida real:** Es la herramienta principal para la fase de "Limpieza de Datos" (Data Wrangling) antes de hacer cualquier gráfico.

## **3. Matplotlib (El lienzo del artista)**

Es la librería de visualización más antigua y robusta de Python. Te da un control total (y a veces complejo) sobre cada detalle del gráfico.

- **Para qué sirve:** Crear gráficos desde cero (líneas, barras, dispersión). Puedes modificar cada eje, color, etiqueta o fuente.

- **En la vida real:** Se usa cuando necesitas un gráfico muy específico o personalizado que requiere un ajuste fino de los elementos visuales.

## **4. Seaborn (El diseñador de interiores)**

Construida sobre Matplotlib, Seaborn está diseñada para hacer gráficos estadísticos atractivos con muy poco código.

- **Para qué sirve:** Crear visualizaciones complejas de forma sencilla (como mapas de calor o distribuciones estadísticas) con una estética moderna por defecto.

- **En la vida real:** Se usa para la Exploración de Datos (EDA), ya que permite ver patrones y correlaciones de forma visualmente agradable y rápida sin pelearse con el código base de Matplotlib.

## **¿Cómo trabajan juntas? (El flujo de trabajo)**

Imagina que quieres analizar las ventas de una tienda:

**1. Pandas:** Cargas el archivo de ventas y eliminas los datos que están vacíos.

**2. NumPy:** Calculas la media móvil o aplicas una fórmula matemática a los precios.

**3. Seaborn:** Creas rápidamente un gráfico para ver la relación entre publicidad y ventas.

**4. Matplotlib:** Ajustas el tamaño final del gráfico, le pones el logo de la empresa y guardas la imagen en alta resolución.
